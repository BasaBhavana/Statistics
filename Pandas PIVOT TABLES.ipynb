```python
pivot_table1=df.pivot_table(index='DEPTNO',columns='JOB',values='SAL',aggfunc='mean')
pivot_table1                        
```

```python
pivot_table3=df.pivot_table(index='DEPTNO',columns='JOB',values='SAL',aggfunc='sum')
pivot_table3   
```
