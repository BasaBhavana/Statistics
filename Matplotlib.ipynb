```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
marks=[23,44,55,34,33,99]
subjects=['eng','chem','phy','hindi','telugu','science']
```

```python
plt.plot(subjects,marks)
```

```python
plt.figure(figsize=(4,2))
plt.plot(marks,subjects,color='red')
```

```python
plt.figure(figsize=(4,2))
plt.plot(marks,subjects,color='red',linewidth=4)
```

```python
plt.figure(figsize=(4,2))
plt.plot(marks,subjects,color='red',linestyle='dotted')
```

```python
plt.plot(marks,subjects,'bo',linestyle='dashed')
```

```python
x=[10,4,31,14,11,6,12,18,19,10]
y=[100,999,888,700,650,555,488,352,200,100]
```

```python
plt.figure(figsize=(4,2))
plt.xlabel('Days')
plt.ylabel('Price')
plt.title('Stocks')
plt.plot(x,y)
```

```python
plt.figure(figsize=(4,2))
plt.xlabel('Days')
plt.ylabel('Price')
plt.title('Stocks')
plt.plot(x,y,'r+--')
```
