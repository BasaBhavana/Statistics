# Bar charts


```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
shares=['apple','samsung','x','fb','amazon']
price=[1500,1300,800,200,1000]
hike=[price[0]*0.1,price[1]*0.2,price[2]*0.2,price[3]*0.5,price[4]*0.1]
```

```python
ypos=np.arange(len(shares))
ypos
```

```python
plt.figure(figsize=(4,3))
plt.xticks(ypos,shares)
plt.ylabel("Price($)")
plt.title("Tech Company Shares")
plt.bar(ypos-0.1,price,width=0.2,label="Prices")
plt.bar(ypos+0.1,hike,width=0.2,label="Hike")
plt.legend()
```

```python
xpos=np.arange(len(shares))
xpos
```

```python
plt.figure(figsize=(4,3))
plt.xticks(xpos,shares)
plt.title("Tech Company Shares")
plt.bar(xpos-0.1,price,width=0.2,label="Prices")
plt.bar(xpos+0.1,hike,width=0.2,label="Hike")
plt.legend()
```

# Histogram

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
students=[45,67,34,78,43,66,75,89,99]
plt.hist(students)
```

```python
students=[45,67,34,36,54,55,78,43,66,75,89,99]
plt.figure(figsize=(5,3))
plt.hist(students,bins=5,color="skyblue",edgecolor="red",rwidth=0.75)
plt.title("Students Scores")
plt.xlabel("Range of scores")
plt.ylabel("No of students")
plt.show()
```
