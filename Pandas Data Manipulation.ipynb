```python
data['gross_sal']=data['SAL'].transform(lambda n:n+(n*0.45)+(n*0.35)+(n*0.25))
```

```python
my_data=data
my_data
```

```python
result=(
    my_data[my_data['SAL']>1500]
    .sort_values('SAL',ascending=False)
    .groupby('DEPTNO')
    .apply(lambda g:g.assign(gross_sal=g['SAL']+100))
)


#here apply function works on each dept_no/emp_no separately
#which is useful if the transformation depends on group-specific calculations
```
